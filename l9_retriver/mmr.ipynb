# Maximum Marginal Relevance (MMR)

> **"How can we pick results that are not only relevant to the query but also different from each other?"**

---

## 📌 What is MMR?

**MMR (Maximum Marginal Relevance)** is an information retrieval algorithm designed to:

- Reduce **redundancy** in retrieved results  
- Maintain **high relevance** to the query  

---

## 🤔 Why MMR Retriever?

In regular similarity search, you may get documents that are:

- Very similar to each other  
- Repeating the same information  
- Lacking diverse perspectives  

---

## ✅ How MMR Solves This

MMR Retriever avoids redundancy by:

1. Picking the **most relevant document first**
2. Then selecting the next document that is:
   - Still relevant  
   - But **least similar** to already selected documents  
3. Repeating this process for remaining documents  

---

## 🚀 Where is MMR Useful?

Especially useful in **RAG (Retrieval-Augmented Generation)** pipelines:

- When you want your context window to contain  
  **diverse but still relevant information**
- When documents are **semantically overlapping**

---

## 🧠 Key Idea

> Balance between **relevance** and **diversity**

```python
from langchain_chroma import Chroma
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_core.documents import Document
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
embeddings = GoogleGenerativeAIEmbeddings(
    model="models/gemini-embedding-001"
) 
```

```python
# Sample documents
documents = [
    Document(page_content="LangChain makes it easy to work with LLMs."),
    Document(page_content="LangChain is used to build LLM based applications."),
    Document(page_content="Chroma is used to store and search document embeddings."),
    Document(page_content="Embeddings are vector representations of text."),
    Document(page_content="MMR helps you get diverse results when doing similarity search."),
    Document(page_content="LangChain supports Chroma, FAISS, Pinecone, and more."),
]
```

```python
vector_store=Chroma.from_documents(
    documents=documents,
    embedding=embeddings,
    collection_name="mmr",
    persist_directory="my-chroma"
)
```

```python
retriver=vector_store.as_retriever(
    search_type="mmr",
    search_kwargs={"k":3,"lambda_mult":0.5}
)

# k=No of top Results and lambda_mult relevance diversity balance
```

```python
query="What is Langchain?"
results=retriver.invoke(query)
```

```python
for i,doc in enumerate(results):
    print(f"\n ---Result {i+1} ---")
    print(f"Content:- \n {doc.page_content}")
```
