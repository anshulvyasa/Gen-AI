```python
from langchain_chroma import Chroma
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_core.documents import Document
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
embeddings = GoogleGenerativeAIEmbeddings(
    model="models/gemini-embedding-001"
) 
```

```python
# Defining The Documents
documents = [
    Document(page_content="LangChain helps developers build LLM applications easily."),
    Document(page_content="Chroma is a vector database optimized for LLM-based search."),
    Document(page_content="Embeddings convert text into high-dimensional vectors."),
    Document(page_content="OpenAI provides powerful embedding models."),
]
```

```python
vector_store=Chroma.from_documents(
    documents=documents,
    embedding=embeddings,
    collection_name="omega",
    persist_directory="my-chroma"
)
```

```python
# we can also retrive document directly from vector_store but it only provide one stragety where as this retriver 
# that we are creating can use different stragety 
# One More benefit is retrive is a Runnable

retriver=vector_store.as_retriever(search_kwargs={"k":2})
```

```python
query="What is Chroma used for"
results=retriver.invoke(query)
```

```python
for i,doc in enumerate(results):
    print(f"\n--- Result {i+1} ----")
    print(f"Content:- \n {doc.page_content}")
```
