```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict 
```

```python
class BatsmanState(TypedDict):
    runs:int
    balls:int
    fours:int
    sixes:int   
    strike_rate:float
    balls_per_boundry:float
    boundary_percentage:float
    summary:str
```

```python
def calc_strikerate(state:BatsmanState)->BatsmanState:
    strike_rate=(state['runs']/state['balls'])*100
    state['strike_rate']=strike_rate
    
    return {'strike_rate':strike_rate}

def calculate_ball_per_boundary(state:BatsmanState)->BatsmanState:
    bpb=state['balls']/(state['fours']+state['sixes'])
     
    state['balls_per_boundry']=bpb
    return {'balls_per_boundry':bpb}

def calculate_boundary_percentage(state:BatsmanState)->BatsmanState:
    boundary_percent=((state['fours']+state['sixes'])/state['balls'])*100
    
    state['boundary_percentage']=boundary_percent
    return {'boundary_percentage':boundary_percent}

def summary(state:BatsmanState)->BatsmanState:
    summary=f"""
    Strike_Rate - {state["strike_rate"]} \n 
    Ball Per Boundary - {state['balls_per_boundry']} \n 
    Boundary Percentage - {state['boundary_percentage']}
    """
    
    state["summary"]=summary
    
    return {"summary":summary}
```

```python
# Graph
graph=StateGraph(BatsmanState)

# Add Nodes
graph.add_node('calc_strikerate',calc_strikerate)
graph.add_node('calculate_ball_per_boundary',calculate_ball_per_boundary)
graph.add_node('calculate_boundary_percentage',calculate_boundary_percentage)
graph.add_node('summary',summary)


# Add Edges
graph.add_edge(START,'calc_strikerate')
graph.add_edge(START,'calculate_ball_per_boundary')
graph.add_edge(START,'calculate_boundary_percentage')

graph.add_edge('calc_strikerate','summary')
graph.add_edge('calculate_ball_per_boundary','summary')
graph.add_edge('calculate_boundary_percentage','summary')

graph.add_edge('summary',END)


# Compile
workflow=graph.compile()
```

```python
initialState={
    'runs':200,
    'balls':100,
    'fours':12,
    'sixes':16
}

final_state=workflow.invoke(initialState)
print(final_state)
```
