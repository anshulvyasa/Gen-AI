```python
from langchain_core.tools import tool
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import HumanMessage
import requests

load_dotenv()
```

```python
@tool
def get_conversion_factor(base_currency:str,target_currency:str)->float:
    """ This Function Fetches The Current Factor of Coversion From base to Target Currency"""
    
    url = f'https://v6.exchangerate-api.com/v6/95dbc3e7046be23a22af2f1e/pair/{base_currency}/{target_currency}'
    response=requests.get(url)
    
    return response.json()


def convert(base_currency_value:int,conversion_rate:float)->float:
    """Given a Conversion Rate this function will return The target Currency after multiplying it with the base_currency"""
    
    return base_currency_value*conversion_rate
```

```python
res=get_conversion_factor.invoke({'base_currency':'USD','target_currency':'INR'})
print(res)
```

```python
llm=ChatGoogleGenerativeAI(model="gemini-3.1-flash-lite-preview")
llm_with_tools=llm.bind_tools([get_conversion_factor,convert])
```

```python
messages=[HumanMessage("What is the conversion factor between INR and USD, and based on that can you convert 10 inr to usd using the second tool i gave you i.e convert")]
```

```python
ai_message=llm_with_tools.invoke(messages)
```

```python
messages.append(ai_message)
```
