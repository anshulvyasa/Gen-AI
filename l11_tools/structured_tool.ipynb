```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel,Field
```

```python
class MultiplyInput(BaseModel):
    a:int = Field(required=True,description="The First Input To The Function")
    b:int = Field(required=True,description="The Second Input To The Function")
```

```python
def multiply(a:int,b:int)->int:
    return a*b
```

```python
multiply_tool=StructuredTool.from_function(
    func=multiply,
    description="This Multiple Two Number",
    name="multiply",
    args_schema=MultiplyInput
)
```

```python
result=multiply_tool.invoke({"a":3,"b":5})

print(result)
print(multiply_tool.name)
print(multiply_tool.description)
```
