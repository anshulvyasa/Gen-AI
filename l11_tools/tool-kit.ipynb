```python
from langchain_core.tools import tool
```

```python
@tool
def add(a:int,b:int)->int:
    """" Adding Two Number"""
    return a+b

@tool
def mul(a:int,b:int)->int:
    """Multiplying Two Number"""
    return a*b
```

```python
class MathToolKit:
    def get_tools(self):
        return [add,mul]
```

```python
toolkit=MathToolKit()
tools=toolkit.get_tools()

for tool in tools:
    print(tool.name," => ",tool.description)
```
