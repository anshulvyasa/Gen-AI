```python
from langchain_core.tools import tool
```

```python
@tool
def multiple(a:int,b:int)->int:
    """ Multiply Two Number """
    return a*b
```

```python
mul_res=multiple.invoke({"a":5,"b":3})
```

```python
print(mul_res)
```

```python
print(multiple.name)
print(multiple.description)
print(multiple.args)
```
